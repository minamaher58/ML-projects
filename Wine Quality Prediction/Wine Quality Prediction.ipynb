## Data Preprocessing

```python
X = wine_dataset.drop('quality', axis=1)
```

```python
Y = wine_dataset['quality'].apply(lambda y_value: 1 if y_value >= 6 else 0)
```

## Train Test split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```
